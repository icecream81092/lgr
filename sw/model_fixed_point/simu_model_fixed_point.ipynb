```python
def floatToTwosComplementDecimal(intBits,decBits,number):
    if decBits == 0:
        mx = pow(2,intBits-1) - 1 # maximum number
    else:
        mx = pow(2,intBits-1) - pow(2,-1*decBits) # maximum number
    mn = -1*pow(2,intBits-1) # minimum number
    if number > mx:
        print ("number:" + str(number) + " has been truncated to: " + str(mx))
        number = mx
    elif number < mn:
        print ("number:" + str(number) + " has been truncated to: " + str(mn))
        number = mn
    n = []
    m = 0
    if number < 0:
        n.append(1)
        m = -1*pow(2,intBits-1)
    else:
        n.append(0)
        m = 0
    for i in reversed(range(intBits-1)):
        m1 = m + pow(2,i)
        if number < m1:
            n.append(0)
        else:
            n.append(1)
            m = m1
    for i in range(1,decBits+1):
        m1 = m + pow(2,-1*i)
        if number < m1:
            n.append(0)
        else:
            n.append(1)
            m = m1
    return "".join(str(i) for i in n)
print('You have successfully defined the floatToTwosComplementDecimal function!')

def twosComplementDecimalToFloat(intBits,decBits,binString):
    n = 0.0
    if binString[0] == "1":
        n = -1*pow(2,intBits-1)
    for i in range(intBits-1):
        n += int(binString[intBits-1-i])*pow(2,i)
    for i in range(1,decBits+1):
        n += int(binString[intBits-1+i])*pow(2,-1*i)
    return n
print('You have successfully defined the twosComplementDecimalToFloat function!')

def float_to_hexadecimal(number):
    intBits=4
    decBits=4
    if decBits == 0:
        mx = pow(2,intBits-1) - 1 # maximum number
    else:
        mx = pow(2,intBits-1) - pow(2,-1*decBits) # maximum number
    mn = -1*pow(2,intBits-1) # minimum number
    if number > mx:
#         print ("number:" + str(number) + " has been truncated to: " + str(mx))
        number = mx
    elif number < mn:
#         print ("number:" + str(number) + " has been truncated to: " + str(mn))
        number = mn
    n = []
    m = 0
    if number < 0:
        n.append(1)
        m = -1*pow(2,intBits-1)
    else:
        n.append(0)
        m = 0
    for i in reversed(range(intBits-1)):
        m1 = m + pow(2,i)
        if number < m1:
            n.append(0)
        else:
            n.append(1)
            m = m1
    for i in range(1,decBits+1):
        m1 = m + pow(2,-1*i)
        if number < m1:
            n.append(0)
        else:
            n.append(1)
            m = m1
    return hex(int("".join(str(i) for i in n),2))[2:].zfill(2)
print('You have successfully defined the float_to_hexadecimal function!')
```

```python
import math
import numpy as np

def get_param(path):
    file = open(path,'r')
    line = file.readline()
    buf = np.empty(0,int)
    while line:
        buf =np.append(buf,line)
        line = file.readline()
    file.close
    return(int(buf[1][0:5],16),int(buf[2][0:5],16),int(buf[3][0:5],16),int(buf[4][0:5],16),int(buf[5][0:5],16),int(buf[6][0:5],16),int(buf[7][0:5],16),int(buf[8][0:5],16),int(buf[9][0:5],16),int(buf[10][0:5],16),int(buf[11][0:5],16),int(buf[12][0:5],16))
print('You have successfully defined the get_param function!')

def get_x_paddingsame(path):
    x = np.empty((0,30),float)
    padding_row = 30*str(16*'0' + ' ')
    padding_row = padding_row.split(' ')[0:-1]
    x = np.append(x,np.expand_dims(padding_row, axis=0),axis=0)

    row = np.array([])
    xf = open(path,'r')
    line = xf.readline()
    while line:
        row = np.append(row,16*'0')
        for i in range(28): 
            value = "{0:08b}".format(int(line, 16))
            value = 8*value[0]+value
            row = np.append(row,value)
            line = xf.readline()
        row = np.append(row,16*'0')
        x = np.append(x,np.expand_dims(row, axis=0),axis=0)
        row = np.array([])
    xf.close()
    
    x = np.append(x,np.expand_dims(padding_row, axis=0),axis=0)
    return x
print('You have successfully defined the get_x_paddingsame function!')
def get_w(path, K):
    # K is kernel size
    w = np.empty((0,K,K),float)
    w_file = open(path,'r')
    line = w_file.readline()
    while line:
        kernel = np.empty((0,K),float)
        for i in range(K):
            row = np.array([])
            for j in range(K): 
                value = "{0:08b}".format(int(line, 16))
                value = 8*value[0]+value
                row = np.append(row,value)
                line = w_file.readline()
            kernel = np.append(kernel,np.expand_dims(row, axis=0),axis=0)
        w = np.append(w,np.expand_dims(kernel, axis=0),axis=0)
    w_file.close()
    return w
print('You have successfully defined the get_w function!')
def get_b(path,M):
    b = np.empty((M,0),float)
    b_file = open(path,'r')
    line = b_file.readline()
    while line:
        value = "{0:08b}".format(int(line, 16))
        value = 8*value[0]+value
        line = b_file.readline()
        b = np.append(b,value)
    b_file.close()
    return b
print('You have successfully defined the get_b function!')
```

```python
# define function
def single_conv(x_input, weight,bias):
    # x_input 3x3
    # weight 3x3
    # output 1x1
    output = 0
    convo_f = 0
    K = x_input.shape[0]
    for i in range(K):
        for j in range(K):
            product_0 = int(x_input[i, j], 2) * int(weight[i, j], 2)
            product_0 = twosComplementDecimalToFloat(8, 8, bin(product_0)[2:].zfill(16)[-16:])
            input_f = twosComplementDecimalToFloat(4, 4, x_input[i, j][-8:])
            weight_f = twosComplementDecimalToFloat(4, 4, weight[i, j][-8:])
            assert (float_to_hexadecimal(product_0) == float_to_hexadecimal(input_f * weight_f))
            convo_f = convo_f + input_f * weight_f
            product_0 = floatToTwosComplementDecimal(8, 8, product_0)
            output = floatToTwosComplementDecimal(8, 8, output)
            output = int(output, 2) + int(product_0, 2)
            output = twosComplementDecimalToFloat(8, 8, bin(output)[2:].zfill(16)[-16:])
            assert (floatToTwosComplementDecimal(8, 8, output) == floatToTwosComplementDecimal(8, 8,convo_f))
    output = floatToTwosComplementDecimal(8, 8, output)
    output = int(output[4:12], 2) + int(output[-4]) + int(bias, 2)
    output = twosComplementDecimalToFloat(4, 4, bin(output)[2:].zfill(8)[-8:])
    output = max(output, 0)
    return output
print('You have successfully defined the single_conv function!')
```

```python
def feature_map(x_input, weight,bias):
    R = x_input.shape[0] - 2 # substract padding - should input this parameter
    C = x_input.shape[1] - 2 # substract padding - should input this parameter
    K = weight.shape[0]
    output = np.empty((0,C), float)
    for rr in range(0,R):
        width = np.empty(0, float)
        for cc in range(0,C):
            single_sum = single_conv(x_input[rr:rr + K,cc:cc + K],weight,bias)
            width = np.append(width, np.expand_dims(single_sum, axis=0), axis=0)
        output = np.append(output, np.expand_dims(width, axis=0), axis=0)
    return output
print('You have successfully defined the feature_map function!')
```

```python
# Load data for next conv2
def get_x(path):
    R=16
    C=16
    N=16
    DATA_SIZE=8
    x = np.empty((0,R,C),float)
    x_file = open(path,'r')
    line = x_file.readline()
    while line:
        kernel = np.empty((0,C),float)
        for i in range(R):
            row = np.array([])
            for j in range(C): 
                value = "{0:08b}".format(int(line, 16))
                value = 8*value[0]+value
                row = np.append(row,value)
                line = x_file.readline()
            kernel = np.append(kernel,np.expand_dims(row, axis=0),axis=0)
        x = np.append(x,np.expand_dims(kernel, axis=0),axis=0)
    x_file.close()
    return x
print('You have successfully defined the get_x_paddingsame function!')

def get_w_conv2(path,N,K):
    w = np.empty((0,N,K,K),float)
    w_file = open(path,'r')
    line = w_file.readline()
    while line:
        kernel_with_depth = np.empty((0,K,K),float)
        for i in range(N):
            kernel = np.empty((0,K),float)
            for i in range(K):
                row = np.array([])
                for j in range(K): 
                    value = "{0:08b}".format(int(line, 16))
                    value = 8*value[0]+value
                    row = np.append(row,value)
                    line = w_file.readline()
                kernel = np.append(kernel,np.expand_dims(row, axis=0),axis=0)
            kernel_with_depth = np.append(kernel_with_depth, np.expand_dims(kernel, axis=0), axis=0)
        w = np.append(w,np.expand_dims(kernel_with_depth, axis=0),axis=0)
    w_file.close()
    return w
print('You have successfully defined the get_w function!')
```

```python
# define function
def single_conv_2(x_input, weight,bias):
    N = weight.shape[0]
    K = x_input.shape[1]
    output_sum = 0
    for n in range(N):
        output = 0
        convo_f = 0
        for i in range(K):
            for j in range(K):
                product_0 = int(x_input[n, i, j], 2) * int(weight[n, i, j], 2)
                product_0 = twosComplementDecimalToFloat(8, 8, bin(product_0)[2:].zfill(16)[-16:])
                input_f = twosComplementDecimalToFloat(4, 4, x_input[n, i, j][-8:])
                weight_f = twosComplementDecimalToFloat(4, 4, weight[n, i, j][-8:])
                assert (float_to_hexadecimal(product_0) == float_to_hexadecimal(input_f * weight_f))
                convo_f = convo_f + input_f * weight_f
                product_0 = floatToTwosComplementDecimal(8, 8, product_0)
                output = floatToTwosComplementDecimal(8, 8, output)
                output = int(output, 2) + int(product_0, 2)
                output = twosComplementDecimalToFloat(8, 8, bin(output)[2:].zfill(16)[-16:])
                assert (floatToTwosComplementDecimal(8, 8, output) == floatToTwosComplementDecimal(8, 8,convo_f))
        output = floatToTwosComplementDecimal(8, 8, output)
        output_sum = floatToTwosComplementDecimal(8, 8, output_sum)
        output_sum = int(output_sum, 2) + int(output, 2)
        output_sum = twosComplementDecimalToFloat(8, 8, bin(output_sum)[2:].zfill(16)[-16:])
    output_sum = floatToTwosComplementDecimal(8, 8, output_sum)
    output_sum = int(output_sum[4:12], 2) + int(output_sum[-4]) + int(bias, 2)
    output_sum = twosComplementDecimalToFloat(4, 4, bin(output_sum)[2:].zfill(8)[-8:])
    output_sum = max(output_sum, 0)
    return output_sum

print('You have successfully defined the single_conv_2 function!')

def feature_map_3d(x_input, weight, bias):
    R = x_input.shape[1] - 2 # substract padding - should input this parameter
    C = x_input.shape[2] - 2 # substract padding - should input this parameter
    K = weight.shape[1]
    output = np.empty((0,C), float)
    for rr in range(0,R):
        width = np.empty(0, float)
        for cc in range(0,C):
            single_sum = single_conv_2(x_input[:,rr:rr + K,cc:cc + K],weight,bias)
            width = np.append(width, np.expand_dims(single_sum, axis=0), axis=0)
        output = np.append(output, np.expand_dims(width, axis=0), axis=0)
    return output
```

```python
# max pooling for without padding
def max_pooling(x_input):
    R = C = x_input.shape[0]
    mR = mC = R//2
    maxp0 = -1
    r0 = np.array([])
    l1_out0 = np.empty((0,mR),float)
    for i in range(0,R-1,2):
        r0 = np.array([])
        for j in range(0,R-1,2):
            for m in range(2):
                for n in range(2):
                    maxp0 = max(maxp0,x_input[i+m,j+n]) 
            r0 = np.append(r0,maxp0)
            maxp0 = -1
        l1_out0 = np.append(l1_out0,np.expand_dims(r0, axis=0),axis=0)
    return l1_out0
```
